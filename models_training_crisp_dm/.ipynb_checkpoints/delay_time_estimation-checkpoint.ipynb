# Delay Time Estimation

## Imports

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, r2_score
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
from xgboost import XGBRegressor, plot_importance
```

## Load Data

```python
data = pd.read_csv('TAS_Airport_Data_2024_balanced.csv')
data
```

## Preprocessing

```python
# Drop flights with missing departure or arrival delay
data = data.dropna(subset=['DepartureDelay', 'ArrivalDelay'])

# Features to use
features = [
    'FlightDuration', 'BaggageHandlers', 'GroundCrew', 'FuelTrucks', 'CleaningCrew',
    'Passengers', 'AircraftType', 'SecurityChecks', 'WeatherCondition', 'PreviousFlightDelay',
    'GateAvailability', 'GateType', 'AircraftTurnaroundTime', 'CheckInQueueTime', 'SecurityWaitTime'
]

X = data[features]

# Target: DepartureDelay and ArrivalDelay
y_dep = data['DepartureDelay']
y_arr = data['ArrivalDelay']

# Encode categorical features
X = pd.get_dummies(X, columns=['AircraftType', 'WeatherCondition', 'GateAvailability', 'GateType'], drop_first=True)
```

## Train/Test Split

```python
X_train_dep, X_test_dep, y_train_dep, y_test_dep = train_test_split(X, y_dep, test_size=0.2, random_state=42)
X_train_arr, X_test_arr, y_train_arr, y_test_arr = train_test_split(X, y_arr, test_size=0.2, random_state=42)
```

## Model Training

```python
model_dep = XGBRegressor(n_estimators=300, learning_rate=0.05, max_depth=7, random_state=42)
model_dep.fit(X_train_dep, y_train_dep)

model_arr = XGBRegressor(n_estimators=300, learning_rate=0.05, max_depth=7, random_state=42)
model_arr.fit(X_train_arr, y_train_arr)
```

## Evaluation

```python

def evaluate_model(model, X_test, y_test, label="Model"):
    preds = model.predict(X_test)
    mae = mean_absolute_error(y_test, preds)
    r2 = r2_score(y_test, preds)
    print(f"{label} - MAE: {mae:.2f}, R2: {r2:.2f}")
    return preds

print("\n--- Delay Time Estimation Evaluation ---")
preds_dep = evaluate_model(model_dep, X_test_dep, y_test_dep, "Departure Delay")
preds_arr = evaluate_model(model_arr, X_test_arr, y_test_arr, "Arrival Delay")
```

```python
import joblib

joblib.dump(model_dep, "trained_models/delay_time_estimation_dep.pkl")
joblib.dump(model_arr, "trained_models/delay_time_estimation_arr.pkl")
```

## Visualization

```python
def plot_actual_vs_predicted(y_true, y_pred, title):
    plt.figure(figsize=(8,6))
    sns.scatterplot(x=y_true, y=y_pred, alpha=0.6)
    plt.plot([y_true.min(), y_true.max()], [y_true.min(), y_true.max()], '--r')
    plt.xlabel('Actual Delay')
    plt.ylabel('Predicted Delay')
    plt.title(title)
    plt.grid(True)
    plt.show()

def plot_residuals(y_true, y_pred, title):
    residuals = y_true - y_pred
    plt.figure(figsize=(8,6))
    sns.histplot(residuals, kde=True, bins=30, color='purple')
    plt.xlabel('Residuals')
    plt.title(title)
    plt.grid(True)
    plt.show()

def plot_feature_importance(model, title):
    plt.figure(figsize=(10,8))
    plot_importance(model, max_num_features=15, height=0.5)
    plt.title(title)
    plt.show()
    
    
print("\n--- Visualizations ---")
# Departure Delay Visuals
plot_actual_vs_predicted(y_test_dep, preds_dep, "Departure Delay: Actual vs Predicted")
plot_residuals(y_test_dep, preds_dep, "Departure Delay Residuals Distribution")
plot_feature_importance(model_dep, "Top Features Influencing Departure Delay")

# Arrival Delay Visuals
plot_actual_vs_predicted(y_test_arr, preds_arr, "Arrival Delay: Actual vs Predicted")
plot_residuals(y_test_arr, preds_arr, "Arrival Delay Residuals Distribution")
plot_feature_importance(model_arr, "Top Features Influencing Arrival Delay")

```
