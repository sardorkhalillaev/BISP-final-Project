```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("simulated_data/TAS_Airport_Data_2024_final.csv")

# Display the first few rows
print(df.head())

# Get information about the columns and their data types
print(df.info())

# Check for missing values
print(df.isnull().sum())
```

```python
# Histograms for key numerical features
df.hist(column=["FlightDuration", "Passengers", "DepartureDelay", 
                "ArrivalDelay", "CheckInQueueTime", "SecurityWaitTime", 
                "AircraftTurnaroundTime"], figsize=(15, 10))
plt.show()

# Box plots for delay distributions
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[["DepartureDelay", "ArrivalDelay"]])
plt.title("Distribution of Departure and Arrival Delays")
plt.show()
```

```python
# Bar plots for categorical features
categorical_cols = ["Airline", "WeatherCondition", "ReasonForDelay", "GateAvailability", 
                    "GateType", "LoungeUsage", "OnTimePerformance"]

for col in categorical_cols:
    plt.figure(figsize=(8, 5))
    df[col].value_counts().plot(kind="bar")
    plt.title(f"Distribution of {col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Scatter plot for Passengers vs. Departure Delay
plt.figure(figsize=(8, 6))
sns.scatterplot(x="Passengers", y="DepartureDelay", data=df)
plt.title("Relationship between Passengers and Departure Delay")
plt.show()

# Correlation matrix and heatmap
corr_matrix = df.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix of Numerical Features")
plt.show()
```

```python
# Box plot for Departure Delay by Airline
plt.figure(figsize=(12, 8))
sns.boxplot(x="Airline", y="DepartureDelay", data=df)
plt.title("Departure Delay by Airline")
plt.xticks(rotation=45)
plt.show()

# Violin plot for Arrival Delay by Weather Condition
plt.figure(figsize=(10, 6))
sns.violinplot(x="WeatherCondition", y="ArrivalDelay", data=df)
plt.title("Arrival Delay by Weather Condition")
plt.show()
```

```python
df["ScheduledTime"] = pd.to_datetime(df["ScheduledTime"])
df.set_index("ScheduledTime", inplace=True)

# Resample to daily averages
daily_delays = df["DepartureDelay"].resample("D").mean() 

plt.plot(daily_delays)
plt.xlabel("Date")
plt.ylabel("Average Departure Delay (mins)")
plt.title("Daily Average Departure Delays Over Time")
plt.show()
```

```python
# Pair plot for selected features
sns.pairplot(df[["FlightDuration", "Passengers", "DepartureDelay", "ArrivalDelay", 
                "CheckInQueueTime", "SecurityWaitTime"]])
plt.show()
```

```python
# Average Departure Delay by Airline and Weather Condition
avg_delay = df.groupby(["Airline", "WeatherCondition"])["DepartureDelay"].mean().unstack()
print(avg_delay)

# Heatmap of average delay
plt.figure(figsize=(10, 6))
sns.heatmap(avg_delay, annot=True, cmap="YlGnBu")
plt.title("Average Departure Delay by Airline and Weather Condition")
plt.show()
```
